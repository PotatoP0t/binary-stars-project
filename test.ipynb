```python
import phoebe, numpy as np
from astropy import units as u
import matplotlib.pyplot as plt
```

```python
b = phoebe.default_binary()

b.set_value_all('teff', 5772)
times = phoebe.linspace(0, 1, 1001)
b.add_dataset('lc', times=times, passband='Johnson:V', dataset='lc_V')
b.add_dataset('lc', times=times, passband='Johnson:B', dataset='lc_B')

b.set_value_all('pblum_mode', 'absolute')
```

```python
b.run_compute()
```

```python
b.plot(show=True, legend=True)
```

```python
flux_v = b.get_value('fluxes@lc_V@model')
flux_b = b.get_value('fluxes@lc_B@model')
```

```python
m_v = -2.5 * np.log10(flux_v)
m_b = -2.5 * np.log10(flux_b)



color_bv = (m_b - m_v)
plt.plot(times,color_bv)
```
